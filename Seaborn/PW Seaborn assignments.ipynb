Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans: 
1. Bar-Plot : It is used to represent data in terms of bars between the x and y axis. x-axis can have discrete categories or range of values and on y-axis consists of the value of that category or range.

2. Scatter-Plot : It is used to represent data in terms points or dots in the visualization space either 2D or 3D. This plot helps us to unerstand the avaialable trend or relation between two random variables on x and y axis. It can also have seperated for more than one groups.

3. Histogram : It is used to represent the frequency of discrete categories in a dataset or range of values.

4. jointplot - It is used to create a multi-panel figure that displays a combination of univariate and bivariate plots. It's often used to visualize the joint distribution between two variables along with their marginal distributions.

5. HeatMap - Heatmaps are used to visualize the correlation between multiple variables in a tabular format. They are particularly useful for showing the strength and direction of relationships in a dataset

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic.isnull().sum()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.boxplot(x='pclass', y='age', data=titanic, palette='Set3')

plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.boxplot(x='pclass', y='age', data=titanic, palette='Set3')

plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Ticket Fare")

plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds
```

```python
diamonds.isnull().sum()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(x='price',data=diamonds,hue='cut')
plt.title("Diamond Cut V/S Price Comparsion")
plt.xlabel("Diamond Price")
plt.ylabel("Price Count")
```
